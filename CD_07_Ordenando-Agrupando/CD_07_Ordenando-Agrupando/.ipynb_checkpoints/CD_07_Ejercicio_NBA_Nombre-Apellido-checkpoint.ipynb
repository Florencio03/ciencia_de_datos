```python
def h_to_met(height): #review this part, you should learn more about define funcios
    # Split the height into feet and inches
    feet, inches = map(int, height.split('-'))
    #1 foot = 0.3048 meters
    feet_to_meters = feet * 0.3048
    #1 inch = 0.0254 meters
    inches_to_meters = inches * 0.0254
    height_in_meters = feet_to_meters + inches_to_meters
    return height_in_meters
```

```python
h_to_met('5-5')
```

```python
Newnba_df['Height'].astype(str)
```

```python
Newnba_df.dtypes
```

```python
Newnba_df['Height(m)']=Newnba_df['Height'].apply(h_to_met)
```

```python
Newnba_df['Weight(kg)']=Newnba_df['Weight'].apply(lambda x : x*0.453592)
```
