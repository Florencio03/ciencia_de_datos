Con los datos del archivo "asaltos_yucatan.csv", mostrar en un mapa la ubicación de los delitos en Mérida, de ser posible muestre como _popup_ la categoría.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
```

```python
#Cargar el archivo datos/asaltos_yucatan.csv en un DataFrame llamado data
# si es necesario usar la opcion encoding = "ISO-8859-1"
data = pd.read_csv('datos/asaltos_yucatan.csv', encoding = "ISO-8859-1")
```

```python
#Mostrar las primeras filas del DataFrame data
data.head()
```

```python
data_m= data[data['Ciudad']=='Mérida']
```

```python
data_m
```

```python
data_m['Categoria'].unique()
```

```python
data_m_rc=data_m[data_m['Categoria'] =='robo casa']
```

```python
data_m_rc
```

```python
data_m_r=data_m[data_m['Categoria']=='asalto']
```

```python
data_m_r
```

```python
mapa = folium.Map(location=[20.9785, -89.6332],zoom_start=13)
mapa
```

```python
data_m_rc.apply(lambda row:folium.Marker(location=[row["lat"], row["long"]], popup='robo casa', tooltip='<strong>Click here to see Popup</strong>',
                                         icon=folium.Icon(color='green',icon='none') ).add_to(mapa), axis=1)

data_m_r.apply(lambda row:folium.Marker(location=[row["lat"], row["long"]], popup='asalto', tooltip='<strong>Click here to see Popup</strong>',
                                        icon=folium.Icon(color='red',icon='none') ).add_to(mapa), axis=1)
```

```python
# Mapa de base
mapa=folium.Map(location=[20.9785, -89.6332],zoom_start=12)
# Agregar marcadores al mapa
folium.Marker(location=[21.04829, -89.64025],popup='FIUADY').add_to(mapa)
folium.Marker(location=[20.96972, -89.62272],popup='<strong>Edificio Central UADY</strong>',tooltip='<strong>Click here to see Popup</strong>').add_to(mapa)
folium.Marker(location=[20.94167,-89.59466],popup='<h3 style="color:green;">Estadio Kukulc&aacute;n</h3>',tooltip='<strong>Click here to see Popup</strong>').add_to(mapa)

mapa
```

```python
folium.Marker(location=[21.04829, -89.64025], popup='FIUADY', icon=folium.Icon(color='red',icon='none') ).add_to(mapa)

folium.Marker(location=[20.96972, -89.62272],popup='<strong>Edificio Central UADY</strong>',
              tooltip='<strong>Click here to see Popup</strong>',
              icon=folium.Icon(color='green',prefix='fa',icon='school-flag')     ).add_to(mapa)
```
